### Filter

#### Quetion: What is the average fare when a group has its average age greater than or less 38?

```python
titanic_class.filter( lambda x : x["Age"].mean() > 38.0)
```

### Transform

#### Question: Which of the passengers paid a fare that is higher than the average for their ticket class?

```python
titanic["avg_fare_class"] = titanic.groupby("Pclass")["Fare"].transform(lambda x : x.mean())
```
